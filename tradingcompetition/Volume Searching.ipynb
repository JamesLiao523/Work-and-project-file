### Set Ticker & Date

```python
ticker_lists = "NFLX,NVDA,BABA,BIDU,TWTR,TSLA,AAPL,AMZN,ENPH,FB,GOOG,FSLR,SEDG,AXP,BEN,ALXN,AMGN,ETN"
ranges = ticker_lists.split(",")
start_date = "2015-2-20"
```

```python
term = [5,10,20]
volume_dict={}
for i in ranges:
    Datas = data.get_data_yahoo(i, start_date)
    volume_df = Datas["Volume"]
    volume_df = volume_df.sort_index(ascending=False)
    for n in term:
        name = str(i) + "_" + str(n)
        volume_dict[name] = np.mean(volume_df[:n+1])
volume_dict
```

```python
import csv
col = [["Ticker","volume"]]
with open("volume_results.csv","w", newline='') as f:    #parameter "newline" defines whether leaving space after writing or not
    writer = csv.writer(f)
    writer.writerows(col)
    for key in volume_dict:
        f.write("%s,%.2f\n" %(key, volume_dict[key]))
    f.close()
```
