## Web crawling

```python
import os
from requests.api import options
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from lxml import html
from bs4 import BeautifulSoup
from amazoncaptcha import AmazonCaptcha


def fuckamazon(asin):
    #selenium設定引數
    options = Options()
    #設定無頭引數,即不開啟瀏覽器

    options.add_argument('--headless')
    #設定Chrome瀏覽器的selenium驅動 

    chromedriver="C:/Users/j4a3m/Downloads/chromedriver_win32/chromedriver.exe"
    os.environ["webdriver.chrome.driver"] = chromedriver
    #將引數設定+瀏覽器驅動組合

    browser = webdriver.Chrome(chromedriver,chrome_options=options)
    url= "https:/www.amazon.com/dp/" + str(asin)
    print(url)
    
    
    #通過selenium來存取亞馬遜 + 繞過驗證碼程序
    
    browser.get(url)
    captcha = AmazonCaptcha.fromdriver(browser)
    
    fw=open('C:/Users/j4a3m/Downloads/chromedriver_win32/amzon.html','w',encoding='utf-8')
    fw.write(str(browser.page_source))
    browser.close()
    fw.close()
    
    #讀取本地html file
    
    tree = html.parse('C:/Users/j4a3m/Downloads/chromedriver_win32/amzon.html')
    f = open('C:/Users/j4a3m/Downloads/chromedriver_win32/amzon.html', encoding="utf8")     
    soup = BeautifulSoup(f)
    text = soup.title
    Category = [text.contents[0].split(':') for i in text.contents][0]
    f.close()
    
    return Category
```

## Loop until all products get scraped

```python
def Find_Error(Idiot_Amazon):    #Filter out problematic html and assign to a list
    Failed_list = []
    for i in Idiot_Amazon.keys():
        try:
            Idiot_Amazon[i][1]
        except:
            Failed_list.append(i)
    return Failed_list


def Get_Product(Unique_Asin, Idiot_Amazon):     # Get informaton from html file and assign to a dictionary
    for i in Unique_Asin:
        category = fuckamazon(str(i))
        Idiot_Amazon[str(i)] = category
        
    return Idiot_Amazon
     
```

```python
Idiot_Amazon = {}
Unique_Asin = list(df.asin.unique())

# Loop until all products get scraped
while True:
    try:
        Amazon_get = Get_Product(Unique_Asin, Idiot_Amazon)
        Failed_list = Find_Error(Amazon_get)
        Unique_Asin = Failed_list
        
    except:
        break
```
