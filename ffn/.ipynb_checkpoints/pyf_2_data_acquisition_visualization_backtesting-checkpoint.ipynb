### Data Crawler: Listed Company (上市) at TWSE

#### 1. Set destination folder

````python
from os import mkdir                #路徑相關套件
from os.path import isdir

dst = "data"    #原本路徑建立
if not isdir(dst):
    print("Destination folder: ./{0} not found...creating ./{0}".format(dst))
    mkdir(dst)
else:
    print("Destination folder: ./{}...ready".format(dst))
````

#### 2. Access to TWSE by requests
* [urllib.request](https://docs.python.org/3/library/urllib.request.html): extensible library for opening URLs

````python
import requests     
import time      

date = [2020, 2, 12]
date_str = '{0}{1:02d}{2:02d}'.format(date[0], date[1], date[2])      #時間格式設定
url = 'http://www.twse.com.tw/exchangeReport/MI_INDEX'

query_params = {
    'date': date_str,
    'response': 'json',
    'type': 'ALL',
    '_': str(round(time.time() * 1000) - 500)
}

page = requests.get(url, params = query_params)
content = page.json()
````

````python
print(content)
````

#### 3. Create two functions: clean row data and save to files
* Regular expressions (RE)
    * Tutorial on regular expressions: http://www.cs.fsu.edu/~carnahan/cis4930sp17/Lecture8_cis4930.pdf
    * RE tutorial: https://docs.python.org/3/library/re.html
    * https://regexone.com/ (highly recommended!)

````python
import re
import csv

def clean_row(row):
    for index, content in enumerate(row):
        row[index] = re.sub(",", "", content.strip()) # clean commas and spaces
    return row

def record(stock_id, row):
    with open('{}/{}.csv'.format(dst, stock_id), 'a') as f:
        cw = csv.writer(f, lineterminator = '\n')
        cw.writerow(row)
````

#### 4. Save daily trading data for each stock
* For each row in one file, it includes date, share volume, value volume, open price, high price, low price, close price, close-open, number of trades (shown as the listing).

````python
date_str2 = '{0:04d}/{1:02d}/{2:02d}'.format(date[0], date[1], date[2])

for data in content['data9']: 
    sign = '-' if data[9].find('green') > 0 else ''
    row = clean_row([date_str2,        # 日期
                     data[2],          # 成交股數
                     data[4],          # 成交金額
                     data[5],          # 開盤價
                     data[6],          # 最高價
                     data[7],          # 最低價
                     data[8],          # 收盤價
                     sign + data[10],  # 漲跌價差
                     data[3],          # 成交筆數
    ])
    record(data[0].strip(), row)
````

### Another Example: Monthly Sales
* For example, http://mops.twse.com.tw/nas/t21/sii/t21sc03_108_5_0.html.

````python
import pandas as pd
import requests
from io import StringIO
import time

def monthly_report(year, month):
    
    if year > 1990:
        year -= 1911
    
    url = 'https://mops.twse.com.tw/nas/t21/sii/t21sc03_{}_{}.html'.format(year, month)   #年營收 月營收
    if year <= 98:
        url = 'https://mops.twse.com.tw/nas/t21/sii/t21sc03_{}_{}.html'.format(year, month)
    
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    
    r = requests.get(url, headers=headers)
    r.encoding = 'big5'

    dfs = pd.read_html(StringIO(r.text), encoding = 'big-5')

    df = pd.concat([df for df in dfs if df.shape[1] <= 11 and df.shape[1] > 5])
    
    if 'levels' in dir(df.columns):
        df.columns = df.columns.get_level_values(1)
    else:
        df = df[list(range(0,10))]
        column_index = df.index[(df[0] == '公司代號')][0]
        df.columns = df.iloc[column_index]
    
    df['當月營收'] = pd.to_numeric(df['當月營收'], 'coerce')
    df = df[~df['當月營收'].isnull()]
    df = df[df['公司代號'] != '合計']
    
    time.sleep(5)

    return df
````

### Crawler feat. Pandas

#### 三大法人買賣日報表

````python
import os
import requests
import pandas as pd
from datetime import datetime
from io import StringIO

def crawler(date):
    
    r = requests.get('http://www.tse.com.tw/fund/T86?response=csv&date={}&selectType=ALLBUT0999'.format(date))
    df = pd.read_csv(StringIO(r.text), header = 1).dropna(how = 'all', axis = 1).dropna(how = 'any')
    
    df = df.astype(str).apply(lambda s: s.str.replace(',', ''))
    df['stock_id'] = df['證券代號'].str.replace('=', '').str.replace('"', '')
    df['date'] = date
    df = df.set_index(['stock_id', 'date'])
    
    return df.apply(lambda s: pd.to_numeric(s, errors = 'coerce')).dropna(how = 'all', axis = 1)
````

````python
crawler("20201120")
````

#### 季財報
* Target: http://mops.twse.com.tw/mops/web/t163sb04

````python
import requests
import pandas as pd

def financial_statement(year, quarter, type = '綜合損益表'):

    if year >= 1000:
        year -= 1911

    if type == '綜合損益表':
        url = 'https://mops.twse.com.tw/mops/web/ajax_t163sb04'
    elif type == '資產負債查詢彙總表':
        url = 'https://mops.twse.com.tw/mops/web/ajax_t163sb05'
    elif type == '營益分析查詢彙總表':
        url = 'https://mops.twse.com.tw/mops/web/ajax_t163sb06'
    else:
        print('type does not match')
        return None

    r = requests.post(url, {
        'encodeURIComponent':1,
        'step':1,
        'firstin':1,
        'off':1,
        'TYPEK':'sii',
        'year':str(year),
        'season':str(quarter),
    })

    r.encoding = 'utf8'
    dfs = pd.read_html(r.text)
    
    df = dfs[3]
    headers = df.iloc[0]
    df = df.drop(df.index[0])
    df.columns = headers
    df.set_index("公司名稱", inplace = True)
    
    return df
````

#### Example 1: Monte Carlo Simulation for $\pi$

````python
import random

N = 1000

plt.figure(figsize = (8, 8))
for i in range(N):
    
    x = random.uniform(0, 1)
    y = random.uniform(0, 1)
    
    if x ** 2 + y ** 2 < 1:
        plt.plot(x, y, "r.")
    else:
        plt.plot(x, y, "b.")
````

#### Example 3: Implied Volatility in 3D plot
* Data set: https://www.csie.ntu.edu.tw/~d00922011/python/data/sp500_imv.csv

````python
import pandas as pd

df = pd.read_csv('sp500_imv.csv', index_col  = 0)
````

````python
tenor = df.index.values
print(tenor)
moneyness = list(map(float, df.columns.values))
print(moneyness)
````

````python
import numpy as np

X, Y = np.meshgrid(moneyness, tenor)
````

````python
implv = df[df.columns.values][:].values
print(implv)
````

````python
%matplotlib notebook
````

````python
from matplotlib import interactive
from mpl_toolkits import mplot3d
interactive(True)
````

````python
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import LinearLocator, FormatStrFormatter

fig = plt.figure(figsize = (8, 4))
ax = fig.add_subplot(111, projection = '3d')
ax.set_adjustable(adjustable = 'box')
surf = ax.plot_surface(X, Y, implv * 100, cmap = cm.plasma, rstride = 1, cstride = 1, edgecolor = 'none', alpha = 0.5)

ax.zaxis.set_major_locator(LinearLocator(5))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
fig.colorbar(surf, shrink = 0.6, aspect = 20)
ax.view_init(azim = 190, elev = 20)
ax.set_xlabel('Moneyness')
ax.set_ylabel('Tenor (year)')
ax.set_zlabel('Implied Volatility (%)')

plt.savefig('sp500_imv.png', dpi = 300)
plt.show()
````

````python
import ffn

stock2330 = ffn.get('2330.TW:Open, 2330.TW:High, 2330.TW:Low, 2330.TW:Close', start = "2019-11-1")
stock2330.columns = ["open", "high", "low", "close"]
stock2330.head()
````

````python
inc = stock2330.close > stock2330.open
dec = stock2330.close < stock2330.open
````

````python
from math import pi
import pandas as pd
from bokeh.plotting import figure, show, output_file
from bokeh.models import HoverTool

w = 12 * 60 * 60 * 1000 # half day in ms

TOOLS = "pan, crosshair, wheel_zoom, box_zoom, reset, save"

p = figure(x_axis_type = "datetime", tools = TOOLS, plot_width = 1000, title = "2330 Daily Chart")
p.xaxis.major_label_orientation = pi / 4
p.grid.grid_line_alpha = 0.3
p.title.text_font_size = "16pt"

hover_tool = HoverTool(tooltips = [('Date', '@Date{%F}'), ('Close Price', '@close{%.0,0.2} TWD')],  
                       formatters = {'Date':'datetime', 'close':'printf'})
p.add_tools(hover_tool)

p.segment(stock2330.index, stock2330.open, stock2330.index, stock2330.low, color = "black")
p.vbar(stock2330.index[inc], w, stock2330.open[inc], stock2330.close[inc], fill_color = "#F2583E", line_color = "black")
p.vbar(stock2330.index[dec], w, stock2330.open[dec], stock2330.close[dec], fill_color = "#00cc00", line_color = "black")

output_file("2330.html", title = "Candlestick Plot for Stock 2330")
show(p) 
````

### Example

````python
import pandas as pd
import ffn

s2330 = ffn.get("2330.TW", start = "2019-10-1")
s2330.tail()
````

````python
s2330["SMA5"] = s2330["2330tw"].rolling(5).mean()
s2330["SMA10"] = s2330["2330tw"].rolling(10).mean()

buy_signal_mask = (s2330["SMA5"].shift(2) > s2330["SMA10"].shift(2)) & (s2330["SMA5"].shift(1) < s2330["SMA10"].shift(1))
sell_signal_mask = (s2330["SMA5"].shift(2) < s2330["SMA10"].shift(2)) & (s2330["SMA5"].shift(1) > s2330["SMA10"].shift(1))
````

````python
s2330[["2330tw", "SMA5", "SMA10"]].plot(style = ".-")
s2330["2330tw"][buy_signal_mask].plot(style = "r^", markersize = 10, grid = True)
ax = s2330["2330tw"][sell_signal_mask].plot(style = "gv", markersize = 10, grid = True)
ax.legend(["2330tw", "SMA5", "SMA10", "Buy Signal", "Sell Signal"])
````

````python
in_position = False
s2330["P&L"] = 0
Trades = 0

for i, t in enumerate(s2330["2330tw"].index):
    if not in_position:
        payoff = 0
        if buy_signal_mask[t]:
            in_position = True
            print("Create a long position:", t)
    else:
        payoff = s2330["2330tw"].iloc[i] - s2330["2330tw"].iloc[i - 1]
        if sell_signal_mask[t]:
            in_position = False
            Trades += 1
            print("Close a long position:", t)
    
    s2330["P&L"].iloc[i] = s2330["P&L"].iloc[i - 1] + payoff
    print(t, s2330["P&L"][t])
print("Number of trades", Trades)
````

````python
s2330["DD"] = (s2330["P&L"] - s2330["P&L"].cummax()).cummin()
idx = s2330["DD"].idxmin()
s2330["P&L"].plot(style = ".--")
ax = s2330["DD"].plot(style = ".--", grid = True)
ax.legend()
print("MDD at", idx)
````

### Package: backtrader
* Tutorial: https://www.backtrader.com/docu/introduction.html
    * Data set: [S&P500 Index](https://www.csie.ntu.edu.tw/~d00922011/python/data/^GSPC.csv) (You should use the data of E-Mini S&P 500! See [E-mini S&P 500 Futures Quotes](https://www.cmegroup.com/cn-t/trading/equity-index/us-index/e-mini-sandp500.html).)
        
* The common procedure of backtesting could be conducted as follows:
    * Initiate a Cerebro engine.
    * Load and inject a Data Feed: date, open, high, low, close, volume.
    * Create your strategy using indicators.
    * Run the Cerebro engine.
* Quick starter: https://www.backtrader.com/docu/quickstart/quickstart.html

````python
import backtrader as bt
from datetime import datetime


data = bt.feeds.GenericCSVData(dataname = "./2330.tw.csv",
                               dtformat = ('%Y-%m-%d'),
                               datetime = 0,
                               open = 1,
                               high = 2,
                               low = 3,
                               close = 4,
                               volume = 6)
````

````python
class SimpleStrategy(bt.Strategy):

    def __init__(self):
        
        sma5 = bt.ind.SMA(period = 5)
        sma10 = bt.ind.SMA(period = 10)
        self.crossover = bt.ind.CrossOver(sma5, sma10)
      
    def next(self):
        
        if not self.position:       # not in the market
            if self.crossover > 0:  # if fast crosses slow to the upside
                self.buy()          # enter long                
                
        elif self.crossover < 0:    # in the market & cross to the downside
            self.close()            # close long position
````

````python
class MaxOrderSize(bt.Sizer):
    
    params = (('stake', 1),)
    
    def _getsizing(self, comminfo, cash, data, isbuy):
        
        import math

        if isbuy:
            divide = math.floor(cash / data.close[0])
            self.p.stake = divide
            return self.p.stake

        position = self.broker.getposition(data)

        if not position.size:
            return 0

        return self.p.stake
````

````python
cerebro = bt.Cerebro()
cerebro.adddata(data)
cerebro.addstrategy(SimpleStrategy)
cerebro.addsizer(MaxOrderSize)
cerebro.broker.setcash(10000)
cerebro.broker.setcommission(commission = 0.0015)
````

````python
print("Initial portfolio value:", cerebro.broker.getvalue())
cerebro.run()
````

````python
cerebro.plot()
print("Final portfolio value:", cerebro.broker.getvalue())
````
